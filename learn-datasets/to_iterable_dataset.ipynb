```python
from datasets import Dataset, IterableDataset, load_dataset
```

```python
wikipedia = load_dataset("wikimedia/wikipedia", "20231101.en", split="train")
```

```python
# https://huggingface.co/docs/datasets/process#shard
# Dataset 这种可以random access的是没有shard这个说法的
# 但是他可以轻易的被shard化
print(wikipedia)
```

```python
iterable_wikipedia = wikipedia.to_iterable_dataset()
```

```python
# 可以看到，是有features的
# 但是shard是1
# 所以我们必须正确的设置shard
print(iterable_wikipedia)
```

```python
iterable_wikipedia = wikipedia.to_iterable_dataset(4)
```

```python
# 这样就对啦！
print(iterable_wikipedia)
```

```python
# 但是如果是stream加载一个本来就有shard的dataset呢
streaming_wikipedia = load_dataset("wikimedia/wikipedia", "20231101.en", split="train", streaming=True)
```

```python
# 仍然有features
# 且shard的数量就是数据集原本的shard数量
print(streaming_wikipedia)
```

```python
# https://huggingface.co/docs/datasets/stream#shard
# streaming dataset 也是支持shard的
# 咱们试一下
streaming_wikipedia_shard0 = streaming_wikipedia.shard(num_shards=4, index=0)
streaming_wikipedia_shard1 = streaming_wikipedia.shard(num_shards=4, index=1)
streaming_wikipedia_shard2 = streaming_wikipedia.shard(num_shards=4, index=2)
streaming_wikipedia_shard3 = streaming_wikipedia.shard(num_shards=4, index=3)
```

```python
# 原来如此
print(streaming_wikipedia_shard0)
print(streaming_wikipedia_shard1)
print(streaming_wikipedia_shard2)
print(streaming_wikipedia_shard3)
```
